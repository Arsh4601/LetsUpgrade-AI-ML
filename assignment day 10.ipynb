```python
m=np.random.random((2,2))

m[0][0]=round(m[0][0]*10,2)

m[0][1]=round(m[0][1]*10,2)

m[1][0]=round(m[1][0]*10,2)

m[1][1]=round(m[1][1]*10,2)

print("Original Matrix: ")

print("\n",m)

element_sum= round(m[0][0]+m[0][1]+m[1][0]+m[1][1],2)

column_one=round(m[0][0]+m[1][0],2)

column_two=round(m[0][1]+m[1][1],2)

row_one=round(m[0][0]+m[0][1],2)

row_two=round(m[1][0]+m[1][1],2)

print("\nSum of all elements: ",element_sum)

print("\nSum of each coloumn:")

print("\n    Column1: ",column_one)

print("\n    Column2: ",column_two)

print("\nSum of each row:")

print("\n    row1: ",row_one)

print("\n    row2: ",row_two)
```

```python
m=np.random.random((2,2))

m[0][0]=round(m[0][0]*10,2)

m[0][1]=round(m[0][1]*10,2)

m[1][0]=round(m[1][0]*10,2)

m[1][1]=round(m[1][1]*10,2)


while True:
    
    print("\n")
    
    num=input("Enter a number: ")
    
    try:
        
        
        num=int(num)
        
        break
    
    except:
        
        try:
            
            num=float(num)
            break
        
        except:
            
            print("\nInvalid Entry,renter")
        
        

print("\nThe Matrix is: ")

print("\n",m)


for i in range(0,2):
    
    for j in range(0,2):
        
        print("\nWith respect to ",m[i][j])
        
        sum=round(num+m[i][j],2)
        
        print("\nSum: ",sum)
        
        diff=round(num-m[i][j],2)
        
        print("\nDifference: ",diff)
        
        prod=round(num*m[i][j],2)
        
        print("\nProduct: ",prod)

        
        quotient=round(num/m[i][j],2)
        
        print("\nQuotient: ",quotient)


```

# Q5)

```python
import math

m=np.random.random((2,2))

m[0][0]=round(m[0][0]*10,2)

m[0][1]=round(m[0][1]*10,2)

m[1][0]=round(m[1][0]*10,2)

m[1][1]=round(m[1][1]*10,2)

print("The Matrix is: ")

print("\n",m)


for i in range(0,2):
    
    for j in range(0,2):
        
        num=math.radians(m[i][j])
        sine=round(math.sin(num),2)
        cosine=round(math.cos(num),2)
        tangent=round(math.tan(num),2)
        
        
        print("\n\nSin of ",m[i][j]," is: ",sine)
        print("\nCos of ",m[i][j]," is: ",cosine)
        print("\ntan of ",m[i][j]," is: ",tangent)
        
              

              

```
