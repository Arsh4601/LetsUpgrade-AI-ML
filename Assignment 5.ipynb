```python
str="Hello World"
print("string = ",str)

print("\n1. find()- Used to find the given argument in string,return -1 if argument absent\n")
print("Ans ",str.find("d"))#1
print("\n2. isdigit()- Checks if there are digits in a string\n")
print("Ans ",str.isdigit())#2
print("\n3. count()-  Counts the number of times the argument has occured in the string\n")
print("Ans ",str.count("o"))#3
print("\n4.isalpha()-  Checks if the string has all alphabical characters,return false if parameter absent\n")
print("(In this case the o/p is false since the space between Hello and World is non-alphabetic)\n")
print("Ans ",str.isalpha())#4
print("\n5. islower()- Checks if the string is in lowercase,return false if parameter absent\n")
print("Ans ",str.islower())#5
print("\n6. isupper()-  Checks if the string is in uppercase,return false if parameter absent\n")
print("Ans ",str.isupper())#6
print("\n7. isnumeric() Checks if numeric characters are preent in string,return false if numeric characters absent\n")
print("Ans ",str.isnumeric())#7
print("\n8.capitalize()-  Converts 1st character to uppercase and rest to lowercase ")
print("Ans ",str.capitalize())#8
print("\n9.casefold()-  Makes the string fit for caseless comparisons\n")
print("Ans ",str.casefold())#9
print("\n10. isalnum()-  Checks if the string has alphanumeric characters,if not it returns false\n")
print("Ans ",str.isalnum())#10
print("\n11. endswith()-  Checks if the string ends with the argument,if not it returns false\n")
print("Ans ",str.endswith("d"))#11
print("\n12. index()-  Returns the index at first occurence of the given argument\n")
print("Ans ",str.index("r"))#12
print("\n13. startswith()- Checks if the string starts with the given argument,returns false if not the case\n")
print("Ans ",str.startswith("H"))#13
print("\n14. swapcase()-  Swaps the cases of all characters in string i.e. all uppercase to lowercase and vice versa\n")
print("Ans ",str.swapcase())#14
print("\n15. upper()-  Converts string to uppercase\n")
print("Ans ",str.upper())#15


```
