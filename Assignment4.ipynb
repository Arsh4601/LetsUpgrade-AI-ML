# Q2)

1)Range() is used to return numbers in a given range.Once the upper limit is exceeded it stops executing 

2)It takes 3 parameters namely start,stop and step.

3)Start takes the lower limit of the given range,stop takes up the upper limit of the range,the upperlimit is considered as 1 less than the given limit e.g. range(1,10) would execute 9 times only not 10  and finally step is used to take the number by which it should iterate,its an optional parameter with default value as 1


4)Syntax(start,stop,return)


5)Range() is used in for loop.

6)It is used to execute the for loop until the loop execution has not exceeped the the given rangeOnce the upper limit is exceeded the loop  stops executing 

7)Syntax: for variable in range(start,stop,step)

8)Here is an example:


```python
for i in range(1,11):
    
    print(i)
```
