```python
favour=data[(data["runtime"]>150)& (data["revenue"]>300000000)&(data["homepage"].notna())]
```

```python
favour.shape
```

```python
c=data["vote_average"].mean()

c
```

```python
def imdb(x,m=m,c=c):
    
    v=x["vote_count"]
    
    r=x["vote_average"]
    
    return (v/(v+m)*r)+(m/(m+v)*c)


favour["score"]=favour.apply(imdb,axis=1)
```

```python
favour["score"].head()
```

```python
favour=favour.sort_values("score",ascending=False)
```
